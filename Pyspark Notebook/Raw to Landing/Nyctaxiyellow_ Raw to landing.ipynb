```python
from pyspark.sql.functions import lit, max, min, to_timestamp
```

```python
processed_date = ''
```

**Processing NycTaxi Yellow**

```python
df_yellow = spark.read.format("parquet").load("Files/Raw/Nyctaxiyellow/*")
display(df_yellow)
```

```python
df_yellow.count()
```

```python
# get dataframe row count once (avoid multiple full scans)
row_count = df_yellow.count()

if row_count > 1:
    print("file contains more than one row of data")

    # add a new column "processing_date" with the processed_date value
    df_yellow_new = df_yellow.withColumn("processing_date", lit(processed_date))

    # write the dataframe to the specified path in csv format with partitioning
    df_yellow_new.write.format("csv") \
        .option("header", "true") \
        .mode("append") \
        .partitionBy("processing_date") \
        .save("abfss://NycTaxiProject@onelake.dfs.fabric.microsoft.com/NycTaxiProject.Lakehouse/Files/Landing/NycTaxiyellow")

    print("Data loaded to landing layer successfully")

elif row_count == 0:
    print("file contains no data")

else:
    print("file contains exactly one row")


```
